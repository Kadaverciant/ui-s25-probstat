## Task 1


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import plotly.express as px

def generate_data(N, d):
    X = np.random.randn(N, d)
    return X - np.sum(X, axis=0)/len(X)

def calculate_var(vals):
    mean = sum(vals) / len(vals)
    return sum([(val - mean)**2 for val in vals]) / len(vals)
    
def check_pca(generator, dim=3, N = 10**4, d = 10**2):
    X = generator(N, d)
    cov_matrix = X.T @ X / (N - 1)
    eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)
    
    idx = np.argsort(eigenvalues)[::-1]
    eigenvalues = eigenvalues[idx]
    eigenvectors = eigenvectors[:, idx]
    
    top_eigenvectors = eigenvectors[:, :dim]
    print("Greatest eigenvalues: ", eigenvalues[:dim], "Mean: ", np.sum(eigenvalues)/len(eigenvalues), "Var: ", calculate_var(eigenvalues))
    X_pca = X @ top_eigenvectors
    
    return X_pca

def plot_3_components(data, title, columns):
    fig = px.scatter_3d(pd.DataFrame(data, columns=columns), x=columns[0], y=columns[1], z=columns[2], title=title)
    fig.update_traces(marker_size = 3)
    fig.show()

plot_3_components(check_pca(generate_data), "PCA of N(0, I) for dim=3", ["PC1", "PC2", "PC3"])
```

## Task 2

```python
def generate_data_with_var(N, d):
    A = np.random.rand(d, d)
    Sigma = np.dot(A, A.T)
    X = np.random.multivariate_normal(mean=np.zeros(d), cov=Sigma, size=N)
    return X - np.sum(X, axis=0)/len(X)
    
plot_3_components(check_pca(generate_data_with_var), r"PCA of N(0, ) for dim=3", ["PC1", "PC2", "PC3"])
```

## Results
Eigenvalues for N(0, I) are closer to 1 (and almost the same), when for  N(0, ) they have higher difference between each other

So, we can say, that the reason that for N(0, I) we have sphere is the fact, that variance betweeen eigenvalues is almost 0 and thus each PC is similar to each other. 

At the same time, we can notice very high variance for N(0, ) , which leads to higher magnitude alongside with one of the principal components.

So, PCs which corresponds to similar eigenvalues will have similar magnitude.
